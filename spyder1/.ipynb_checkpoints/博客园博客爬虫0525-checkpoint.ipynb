### 导入包

```python
import requests
from lxml import etree
```

### 获取网页源代码requests

```python
headers={
    "user-agent":"Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.108 Safari/537.36"
}
url="https://www.cnblogs.com/zjc9915"
res=requests.get(url=url,headers=headers).text
print(res)
```

### 解析数据xpath,获得文章URL,获取文章html

```python
dom=etree.HTML(res)
article_url_list=dom.xpath('//a[@class="postTitle2"]/@href')
article_title_list=dom.xpath('//a[@class="postTitle2"]/text()')
# artile={}
# article_title=[]
# article_url=[]
# for article_title in article_title_list:
#     yield article_title
# print(article_title)
for article_url in article_url_list:
    print(article_url)
    html=requests.get(article_url,headers=headers).text
#     print(html.status_code)
    print(html)
    
```
